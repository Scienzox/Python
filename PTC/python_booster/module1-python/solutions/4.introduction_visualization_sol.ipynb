```python
x = np.random.binomial(n=10, p=0.2, size=10000)
z = np.random.normal(loc=10*0.2,scale=np.sqrt(10*0.2*0.8), size=10000)

...
```

#### Solution

```python
x = np.random.binomial(n=10, p=0.2, size=10000)
z = np.random.normal(loc=10*0.2,scale=np.sqrt(10*0.2*0.8), size=10000)

plt.hist(x, bins=np.arange(-5,25,0.5), color="blue", alpha=0.5, density=True);
plt.hist(z, bins=np.arange(-5,25,0.5), color="red", alpha=0.5, density=True);
```

# Advanced options

```python
x = np.linspace(-10, 10, 100)
plt.plot(x,x)
plt.plot(x, -x**2)
plt.axis([-20,20,-120,40])
plt.show()
```

```python
plt.plot(x,x)
plt.plot(x, -x**2)
plt.text(-6, -50, 'hey')
plt.text(-8, 0, r'$y=x$')
```

```python
plt.plot(x,x)
plt.plot(x, -x**2)
plt.annotate('tangent point', xy=(0, 0), xytext=(-5, -40),
             arrowprops=dict(facecolor='black'),
             )
plt.show()
```

```python
plt.plot(x, x**2)
plt.xscale("log")
```

```python
fig = plt.figure(figsize=(10,8))

plt.subplot(221)
plt.plot(x, x)
plt.title('$x$')
plt.grid(True)


plt.subplot(222)
plt.plot(x, x**2)
plt.plot([0], [0], 'ro')
plt.title('$x^2$')
plt.grid(True)

plt.subplot(223)
plt.plot(x, x**3)
plt.plot([0], [0], 'ro')
plt.title('$x^3$')
plt.grid(True)

plt.subplot(224)
plt.plot(x, x**4)
plt.title('$x^4$')
plt.grid(True)
```

## Exercise

* Create a plot of $y(x) = x^3$ in green with xaxis between -10 and 10
* add the tangent to $y(x)$ at $x=10$ as a dashed red line (you will need to remember a bit of maths for this. hint: $\frac{dx^3}{dx} = 3x^2$)
* annotate the point (-5, -125) with some text
* label the axes and choose a convenient title

```python
x = np.linspace(-20, 20, 1000)
tangente = 300*x - 2000

...
```

### Solution

```python
x = np.linspace(-20,20, 1000)
y = x**3

tangente = 300*x - 2000

plt.plot(x, y, c="green", label="$x^3$")
plt.plot(x, tangente, c="red", linestyle = "dashed")

plt.annotate('Here', xy=(-5, -125), xytext=(-10, 4000),
             arrowprops=dict(facecolor='black'),
             );

plt.title("$x^3$ curve with tangent and annotation")
plt.xlabel("x")
plt.ylabel("$y(x) = x^3$");
```

# Plotting time series data

```python
netflix = pd.read_csv("netflix_titles.csv", parse_dates = ["date_added"])
```

```python
netflix.info()
```

```python
timeseries = netflix[["date_added", "type"]].copy()
ts = timeseries.set_index("date_added")
ts.head()
```

```python
ts.resample('M').count().plot(figsize=(12,6))
```

```python
ts.resample('Y').count().plot(figsize=(12,6))
```

```python
timeseries["month"] = timeseries["date_added"].dt.month

pd.crosstab(timeseries.month, timeseries.type).plot(kind="bar", figsize=(12,6))
```
