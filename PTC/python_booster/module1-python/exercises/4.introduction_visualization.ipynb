```python
x = np.linspace(-10, 10, 100)
plt.plot(x, x)
plt.plot(x, x**2)
plt.plot(x, x**3, linewidth=4.0, alpha=0.5)
plt.show()
```

```python
plt.plot(x, x, label="linear")
plt.plot(x, x**2, label="quadratic")
plt.plot(x, x**3, label="cubic")
plt.legend()
plt.show()
```

# Advanced options

```python
x = np.linspace(-10, 10, 100)
plt.plot(x,x)
plt.plot(x, -x**2)
plt.axis([-20,20,-120,40])
plt.show()
```

```python
plt.plot(x,x)
plt.plot(x, -x**2)
plt.text(-6, -50, r'$y=x^2$')
plt.text(-8, 0, r'$y=x$')
```

```python
plt.plot(x,x)
plt.plot(x, -x**2)
plt.annotate('tangent point', xy=(0, 0), xytext=(-5, -40),
             arrowprops=dict(facecolor='black'),
             )
plt.show()
```

```python
plt.plot(x,x)
plt.yscale("log")
```

```python
fig = plt.figure(figsize=(10,8))
plt.subplot(221)
plt.plot(x, x)
plt.title('$x$')
plt.grid(True)


plt.subplot(222)
plt.plot(x, x**2)
plt.plot([0], [0], 'ro')
plt.title('$x^2$')
plt.grid(True)

plt.subplot(223)
plt.plot(x, x**3)
plt.plot([0], [0], 'ro')
plt.title('$x^3$')
plt.grid(True)

plt.subplot(224)
plt.plot(x, x**4)
plt.title('$x^4$')
plt.grid(True)
```

# Plotting time series data

```python
netflix = pd.read_csv("netflix_titles.csv", parse_dates = ["date_added"])
```

```python
netflix.info()
```

```python
timeseries = netflix[["date_added", "type"]].copy()
ts = timeseries.set_index("date_added")
ts.head()
```

```python
ts.resample('M').count().plot(figsize=(12,6))
```

```python
ts.resample('Y').count().plot(figsize=(12,6))
```

```python
timeseries["month"] = timeseries["date_added"].dt.month

pd.crosstab(timeseries.month, timeseries.type).plot(kind="bar", figsize=(12,6))
```
